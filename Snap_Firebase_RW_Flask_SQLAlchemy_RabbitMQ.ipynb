```python
import pandas as pd
from typing import Optional, List

class RecommendationNet():
  nodeDf: str
  edgeDf: str
  nodeField: str
  edgeSrcField: str
  edgeDstField: str
  net: snap.PNEANet
  nodeNumToIdMap: dict
  nodeIdToNumMap: dict

  def __init__(self, 
               nodeDf: pd.DataFrame, 
               edgeDf: pd.DataFrame,
               nodeField: str,
               edgeSrcField: str,
               edgeDstField: str):
    self.nodeDf = nodeDf
    self.edgeDf = edgeDf
    self.nodeField = nodeField
    self.edgeSrcField = edgeSrcField
    self.edgeDstField = edgeDstField
    self.nodeNumToIdMap = dict()
    self.nodeIdToNumMap = dict()
    self.net = snap.PNEANet.New()

  def buildNet(self, nodeMetaData: List[str], edgeMetaData: List[str]) -> snap.PNEANet:
    for index, node in self.nodeDf.iterrows():
      self.nodeIdToNumMap[node[self.nodeField]] = index
      self.nodeNumToIdMap[index] = node[self.nodeField]
      self.net.AddNode(index)
      for field in nodeMetaData: self.net.AddStrAttrDatN(index, node[field], field)

    for index, edge in self.edgeDf.iterrows():
      src, dest = self.nodeIdToNumMap[edge[self.edgeSrcField]], self.nodeIdToNumMap[edge[self.edgeDstField]]
      self.net.AddEdge(src, dest)
      edgeId = self.net.GetEId(src, dest)
      for field in edgeMetaData: self.net.AddStrAttrDatE(edgeId, edge[field], field)
    return self.net

  def getNodeIdToNumMap(self) -> dict:
    if self.net is None: 
      print("Call buildNet first")
    return self.nodeIdToNumMap
  
  def getNodeNumToIdMap(self) -> dict:
    if self.net is None: 
      print("Call buildNet first")
    return self.nodeNumToIdMap

  def printEdgeTable(self):
    print("Loaded Edges:")
    print("-------")
    fields = list(self.edgeDf.columns)
    print(','.join(fields))
    for E in self.net.Edges():
      entry = ""
      for field in fields : entry += self.net.GetStrAttrDatE(self.net.GetEId(E.GetSrcNId(), E.GetDstNId()), field) + " "
      print(entry)
    print("-------")

  def printNodeTable(self):
    print("Loaded Nodes:")
    print("-------")
    fields = list(self.nodeDf.columns)
    print(','.join(fields + ['degree']))
    for N in self.net.Nodes():
      entry = ""
      for field in fields : entry += self.net.GetStrAttrDatN(N.GetId(), field) + " "
      entry += str(N.GetInDeg())
      print(entry)
    print("-------")
```

# **Testing Recommendations**

Python Code/Pseudocode of the Random Walk

```python
def GetRndWalkRestart_PNEANet(SongUserNet: snap.PNEANet,
                              ArtistUserNet: snap.PNEANet,
                              ArtistSongNet: snap.PNEANet,
                              JumpProb: float,
                              RandomHopProb: float,
                              SongUserNetHopThresh: float,
                              SongUserNetStayThresh: float,
                              ArtistUserNetHopThresh: float,
                              ArtistUserNetStayThresh: float,
                              ArtistSongNetHopThresh: float,
                              ArtistSongNetStayThresh: float,
                              StartNIdV: snap.TIntV,
                              Rnd: snap.TRnd,
                              N: int,
                              RwrNIdH: snap.TIntH):

  for i in range(N):
    Graph = SongUserNet
    netType = "Song/User"
    dislikes = 0
    locationId = StartNIdV.GetRndVal(Rnd)
    latestSongId = -1
    steps = 0
    while Rnd.GetUniDev() <= JumpProb:
      steps += 1
      print(netType, "netType")
      #Hop to random node in network
      if Rnd.GetUniDev() <= RandomHopProb:
        print("Random hop")
        locationId = Graph.GetRndNId()
        continue

      #Random neighbor or restart if no neighbor
      location = Graph.GetNI(locationId)
      d = location.GetOutDeg()
      print(d, 'out degree')
      if d > 0:
        locationId = location.GetOutNId(Rnd.GetUniDevInt(d))
      else:
        locationId = StartNIdV.GetRndVal(Rnd)
        continue
      
      #Save latest song id
      if Graph.GetStrAttrDatN(locationId, "type") == "song": 
        latestSongId = locationId
      
      edgeId = Graph.GetEI(location.GetId(), locationId).GetId()
      edgeType = Graph.GetStrAttrDatE(edgeId, "type")
      if (edgeType == "dislike"):
        dislikes+=1
      
      #Network hopping
      hopNet = Rnd.GetUniDev()
      if netType == 'Song/User':
        if hopNet > SongUserNetStayThresh and hopNet > SongUserNetStayThresh and ArtistUserNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Song/User to Artist/User")
          Graph = ArtistUserNet
          netType = "Artist/User"
        elif hopNet > SongUserNetStayThresh and hopNet <= SongUserNetStayThresh and ArtistSongNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Song/User to Artist/Song")
          Graph = ArtistSongNet
          netType = "Artist/Song"
        else:
          Graph = SongUserNet
          netType = "Song/User"
          
      elif netType == "Artist/User":
        if hopNet > ArtistUserNetStayThresh and hopNet > ArtistUserNetStayThresh and SongUserNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Artist/User to Song/User")
          Graph = SongUserNet
          netType = "Song/User"
        elif hopNet > ArtistUserNetStayThresh and hopNet <= ArtistUserNetStayThresh and ArtistSongNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Artist/User to Artist/Song")
          Graph = ArtistSongNet
          netType = "Artist/Song"
        else:
          Graph = ArtistUserNet
          netType = "Artist/User"

      elif netType == "Artist/Song":
        if hopNet > ArtistSongNetStayThresh and hopNet > ArtistSongNetStayThresh and ArtistUserNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Artist/Song to Artist/User")
          Graph = ArtistUserNet
          netType = "Artist/User"
        elif hopNet > ArtistSongNetStayThresh and hopNet <= ArtistSongNetStayThresh and SongUserNet.GetNI(locationId).GetOutDeg() > 0:
          print("Hop Artist/Song to Song/User")
          Graph = SongUserNet
          netType = "Song/User"
        else:
          Graph = ArtistSongNet
          netType = "Artist/Song"

    print(f'Number of steps in RW: {steps}')
    if latestSongId == -1: 
      continue

    score = -1 if dislikes % 2 else 1
    if not RwrNIdH.IsKey(latestSongId):
       RwrNIdH.AddDat(latestSongId, score);
    else:
       RwrNIdH.AddDat(latestSongId, RwrNIdH.GetDat(latestSongId) + score)
```

Wrapper over the random walk implementation (May parameterize other probabilities in future)

```python
def recommendations(userId, jumpProb):
  nodeIdToNumMap = songUserNet.getNodeIdToNumMap()
  nodeNumToIdMap = songUserNet.getNodeNumToIdMap()
  print(f"Recommending for user {userId}")
  StartNIdV = snap.TIntV()
  StartNIdV.Add(nodeIdToNumMap[userId])
  Rnd = snap.TRnd(42)
  RwrNIdH = snap.TIntH()
  snap.GetRndWalkRestart_PNEANet(sunet,
                                 aunet,
                                 asnet,
                                 jumpProb,
                                 0.05,
                                 0.5,
                                 0.5,
                                 0.5,
                                 0.75,
                                 0.75,
                                 0.75,
                                 StartNIdV,
                                 Rnd,
                                 20,
                                 RwrNIdH)
  recommendationsMap = {nodeNumToIdMap[num]: rank for (num, rank) in RwrNIdH.items()}
  #Filter out already liked songs
  likedSongs = edges_users_songs_df.loc[(edges_users_songs_df['userId'] == userId) & (edges_users_songs_df['type'] == 'like')]['songId'].values
  for songId in list(recommendationsMap.keys()):
     if songId in likedSongs: del recommendationsMap[songId]
  return recommendationsMap
```

```python
def consolidateDB(userId, db_records, new_records):
  """
  Consolidation algorithm for existing reccs and new reccs:
  1. If recc with songId exists already, update the score but use existing recommended field
  2. Otherwise add new recc with recommended 'NO'
  3. Re-rank all songs
  4. Update DB
  """
  db_updated = []
  updated_songs = dict()
  db_lookup = {record[1]: index for index, record in enumerate(db_records)}

  for new_record in new_records:
    songId, score = new_record[0], new_record[1]
    if songId in db_lookup:
      updated_songs[songId] = True
      db_record = db_records[db_lookup[songId]]
      recommended = db_record[3] 
      db_updated.append((userId, songId, score, recommended))
    else:
      db_updated.append((userId, songId, score, 'NO'))

  for old_record in db_records:
    songId, score, recommended = old_record[1], old_record[2], old_record[3]
    if songId in updated_songs: continue
    db_updated.append((userId, songId, score, recommended))
  
  return db_updated
```
